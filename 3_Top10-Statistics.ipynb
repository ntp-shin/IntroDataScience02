# **Thống kê top 10 - Top10 Statistics**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
score_df = pd.read_csv('new_data.csv')
score_df
```

```python
provinces = '''
Hà Nội
TP.Hồ Chí Minh
Hải Phòng
Đà Nẵng
Hà Giang
Cao Bằng
Lai Châu
Lào Cai
Tuyên Quang
Lạng Sơn
Bắc Kạn
Thái Nguyên
Yên Bái
Sơn La
Phú Thọ
Vĩnh Phúc
Quảng Ninh
Bắc Giang
Bắc Ninh
Hải Dương
Hưng Yên
Hoà Bình
Hà Nam
Nam Định
Thái Bình
Ninh Bình
Thanh Hoá
Nghệ An
Hà Tĩnh
Quảng Bình
Quảng Trị
Thừa Thiên - Huế
Quảng Nam
Quảng Ngãi
Kon Tum
Bình Định
Gia Lai
Phú Yên
Đắk Lắk
Khánh Hoà
Lâm Đồng
Bình Phước
Bình Dương
Ninh Thuận
Tây Ninh
Bình Thuận
Đồng Nai
Long An
Đồng Tháp
An Giang
Bà Rịa - Vũng Tàu
Tiền Giang
Kiên Giang
Cần Thơ
Bến Tre
Vĩnh Long
Trà Vinh
Sóc Trăng
Bạc Liêu
Cà Mau
Điện Biên
Đăk Nông
Hậu Giang
'''

provinces = provinces.strip().split('\n')
provinces.insert(19, '')

score_df['tinh/tp'] = np.array(provinces)[score_df['cum'] - 1]
score_df
```

## **_01. Top 10 tỉnh có điểm môn Toán >= 9 nhiều nhất:_**

```python
# Gop nhom theo tinh/tp va dem so thi sinh tung tinh
count_math = score_df[score_df['toan'] >= 9].groupby(['tinh/tp'])['tinh/tp'].count()
count_math.sort_values(ascending=False, inplace=True)

# Visualize
plt.title('Top 10 provinces has Math score more than 9.', fontsize=13, fontweight='bold')
plt.xlabel('Total student has more than 9.', fontsize=10)
plt.ylabel('Provinces', fontsize=10)
plt.xlim(0, 4100)

plt.barh(count_math.iloc[:10].index, count_math.iloc[:10].values, color='blue', edgecolor='black')
[plt.text(val + 100, i, str(val)) for i, val in enumerate(np.round_(count_math.iloc[:10].values, 2))];
```

- TP.Hồ Chí Minh và Hà Nội có nhiều điểm Toán >= 9 nhất và cách xa các tỉnh/tp khác vì đây là hai TP lớn có nhiều trường nổi tiếng và học sinh giỏi đồng thời cũng vì là 2 TP đông dân nhất nên số thí sinh chiếm phần lớn.
- Các tỉnh khác như: Nghệ An, Thanh Hóa, Nam Định, Hải Phòng... có nền giáo dục tốt, có nhiều nhân tài và qua các năm đều có tỉ lệ điểm cao hơn.

## **_02. Top 10 tỉnh có điểm môn Ngữ văn >= 9 nhiều nhất:_**

```python
# Gop nhom theo tinh/tp va dem so thi sinh tung tinh
count_lite = score_df[score_df['ngu_van'] >= 9].groupby(['tinh/tp'])['tinh/tp'].count()
count_lite.sort_values(ascending=False, inplace=True)

# Visualize
plt.title('Top 10 provinces has Literature score more than 9.', fontsize=13, fontweight='bold')
plt.xlabel('Total student has more than 9.', fontsize=10)
plt.ylabel('Provinces', fontsize=10)
plt.xlim(0, 3800)

plt.barh(count_lite.iloc[:10].index, count_lite.iloc[:10].values, color='green', edgecolor='black')
[plt.text(val + 100, i, str(val)) for i, val in enumerate(np.round_(count_lite.iloc[:10].values, 2))];
```

- Khác với Toán và Ngoại ngữ, Hải Phòng có số thí sinh trong top này cao nhất, và năm 2022 cũng là lần đầu tiên Hải Phòng có điểm môn văn cao nhất.
- Các tỉnh/tp Hà Nội, Nghệ An, Thanh Hóa, Nam Định, Bắc Ninh,... tiếp tục thuộc top này.

## **_03. Top 10 tỉnh có điểm môn Ngoại ngữ >= 9 nhiều nhất:_**

```python
# Gop nhom theo tinh/tp va dem so thi sinh tung tinh
count_lite = score_df[score_df['ngoai_ngu'] >= 9].groupby(['tinh/tp'])['tinh/tp'].count()
count_lite.sort_values(ascending=False, inplace=True)

# Visualize
plt.title('Top 10 provinces has English score more than 9.', fontsize=13, fontweight='bold')
plt.xlabel('Total student has more than 9.', fontsize=10)
plt.ylabel('Provinces', fontsize=10)
plt.xlim(0, 8500)

plt.barh(count_lite.iloc[:10].index, count_lite.iloc[:10].values, color='yellow', edgecolor='black')
[plt.text(val + 100, i, str(val)) for i, val in enumerate(np.round_(count_lite.iloc[:10].values, 2))];
```

- Hà Nội và TPHCM đứng top đầu, hơn hẳn so với phần còn lại => 2 TP này có chất lượng đào tạo tiếng anh tốt nhất

## **_04. Top 10 tỉnh có tỉ lệ điểm môn Ngoại ngữ >= 5 nhiều nhất:_**

```python
# Gop nhom theo tinh/tp, dem tren tung nhom va sau do tinh ti le tren tong so thi sinh 
count_sum = score_df[~np.isnan(score_df['ngoai_ngu'])].groupby(['tinh/tp'])['tinh/tp'].count()
count_cond = score_df[score_df['ngoai_ngu'] >= 5].groupby(['tinh/tp'])['tinh/tp'].count()
ratio_fore = count_cond / count_sum
ratio_fore.sort_values(ascending=False, inplace=True)

# Visualize
plt.title('Top 10 provinces has Foreign Language score ratio more than 5.', fontsize=13, fontweight='bold')
plt.xlabel('Ratio student has more than 5.', fontsize=10)
plt.ylabel('Provinces', fontsize=10)
plt.xlim(0, .9)

plt.barh(ratio_fore.iloc[:10].index, ratio_fore.iloc[:10].values, color='yellow', edgecolor='black')
[plt.text(val + .03, i, str(val)) for i, val in enumerate(np.round_(ratio_fore.iloc[:10].values, 2))];
```

***Người ta thường nói: Trình độ ngoại ngữ của một vùng sẽ đánh giá mức độ phát triển của vùng đó. Giờ ta sẽ kiểm chứng***

**Điểm Ngoại_ngữ >= 5 tương ứng với trình độ tiếng anh mức độ cơ bản**
- Đứng top đầu là Bình Dương và TPHCM => Đây là nơi được đầu tư môn tiếng anh nhất
- Tiếp đến là Vùng Tàu, Hải Phòng, Đà Nẵng, Hà Nội, Nam Định,... => Các tỉnh thành này đều rất phát triển so với mặt bằng chung. Nhất là về kinh tế, du lịch, cơ sở vật chất - hạ tầng.
- Nghệ An, Hải Dương, Thanh Hóa tuy lọt Top10 tỉnh/thành có số thí sinh trên 9 nhiều nhất nhưng lại không lọt top này. => Chứng tỏ có nhiều người tài giỏi nhưng mặt bằng chung lại không được đầu tư phát triển.

## **_05. Top 10 tỉnh có điểm tổ hợp < 5 nhiều nhất:_**

```python
# Loc ra nhung doi tuong 1 va 2 va co diem to hop < 5 sau do thuc hien gop nhom theo tinh/tp va dem so thi sinh tung tinh
_filter = (np.nanmean(score_df.loc[:, 'vat_li':'gdcd'], axis=1) < 5) & np.in1d(score_df['phan_loai'], ['THPT_KHTN', 'THPT_KHXH', 'GDTX_KHTN', 'GDTX_KHXH'])
count_comb = score_df[_filter].groupby(['tinh/tp'])['tinh/tp'].count()
count_comb.sort_values(ascending=False, inplace=True)

# Visualize
plt.title('Top 10 provinces has Combination Subject score less than 5.', fontsize=13, fontweight='bold')
plt.xlabel('Total student has less than 8.', fontsize=10)
plt.ylabel('Provinces', fontsize=10)
plt.xlim(0, 9700)

plt.barh(count_comb.iloc[:10].index, count_comb.iloc[:10].values, color='red', edgecolor='black')
[plt.text(val + 200, i, str(val)) for i, val in enumerate(np.round_(count_comb.iloc[:10].values, 2))];
```

- Tuy có tỉ lệ điểm cao, nhưng TP.Hồ Chí Minh và Hà Nội vẫn đứng đầu top này vì tỉ lệ thí sinh cao nên dẫn đến số điểm kém (kể cả điểm cao) tăng lên.
- Các tỉnh/tp khác có thể kể đến như: Đắk Lắk, Đồng Nai, Bà Rịa - Vũng Tàu,...

## **_06. Top 10 tỉnh có điểm khối B > 27 nhiều nhất:_**

```python
# Loc ra nhung tat ca doi tuong co diem khoi B > 27 sau do thuc hien gop nhom theo tinh/tp va dem so thi sinh tung tinh
_filter = (np.sum(score_df[['toan', 'hoa_hoc', 'sinh_hoc']], axis=1) > 27)
count_B_comb = score_df[_filter].groupby(['tinh/tp'])['tinh/tp'].count()
count_B_comb.sort_values(ascending=False, inplace=True)

# Visualize
plt.title('Top 10 provinces has B Combination score more than 27', fontsize=13, fontweight='bold')
plt.xlabel('Total student has more than 27.', fontsize=10)
plt.ylabel('Provinces', fontsize=10)
plt.xlim(0, 230)

plt.barh(count_B_comb.iloc[:10].index, count_B_comb.iloc[:10].values, edgecolor='black')
[plt.text(val + 5, i, str(val)) for i, val in enumerate(np.round_(count_B_comb.iloc[:10].values, 2))];
```

**Sở dĩ xét trên 27 vì đa số các ngành y đều phải đạt 27 trở lên mới đậu**
- TPHCM là thành phố có số lượng điểm khối B >= 27 nhiều nhất - top1. => Đây là nơi có nhiều học sinh mong muốn được theo ngành y nhất và cũng một phần do TPHCM đông dân - nhiều trường y (bệnh viện) - Môi trường làm việc tốt. 
- Hà Nội cũng tương tự với top2 nhưng lại không nhiều bằng TPHCM.
- Tiếp đến là Nghệ An, Thanh Hóa, Huế, Đà Nẵng, Cần Thơ => Các thí sinh của các tỉnh/thành này đều có học lực rất tốt.
